```python
final_movie_list = []


for i in data.values.tolist():
    movie_list = []
    
    for j in i:
        if not pd.isnull(j):
            movie_list.append(j)
            
    final_movie_list.append(movie_list)
    
final_movie_list
```

```python
len(final_movie_list)
```

```python
from apyori import apriori
```

```python
basket_intelligence = apriori(transactions=final_movie_list, min_support=0.002, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
results = list(basket_intelligence)
results
```

```python
def inspect (results) :
    product1 = [tuple (result [2][0][0])[0] for result in results]
    product2 = [tuple (result [2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result [2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts) )
DataFrame_intelligence = pd. DataFrame (inspect(results), columns = ['product1', 'product2', 'Support', 'Confidence', 'Lift'])

```
